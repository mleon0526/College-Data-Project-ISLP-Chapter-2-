### Use the plot.hist() method of college to produce some histograms with differing numbers of bins for a few of the quantitative variables.



```python
sns.set_style("whitegrid")

applicationsHistogram = college["Apps"]
acceptanceHistogram = college["Accept"]
expendHistogram = college["Expend"]
graduationGrateHistogram = college["Grad.Rate"]

# calculating the right amount of bins using ScottÂ´s Rule

""" 

ScottÂ´s Rule:

Bin width = 3.5(std of population / n^1/3) 

Number of Bins = Range of data / Bin width

"""

n_applicationsHistogram = len(applicationsHistogram)
n_acceptanceHistogram = len(acceptanceHistogram)
n_expendHistogram = len(expendHistogram)
n_graduationGrateHistogram = len(graduationGrateHistogram)

scott_bins_applicationsHistogram = int((applicationsHistogram.max() - applicationsHistogram.min()) / (3.5 * applicationsHistogram.std() / n_applicationsHistogram**(1/3)))

scott_bins_acceptanceHistogram = int((acceptanceHistogram.max() - acceptanceHistogram.min()) / (3.5 * acceptanceHistogram.std() / n_acceptanceHistogram**(1/3)))

scott_bins_expendHistogram = int((expendHistogram.max() - expendHistogram.min()) / (3.5 * expendHistogram.std() / n_expendHistogram**(1/3)))

scott_bins_graduationGrateHistogram = int((graduationGrateHistogram.max() - graduationGrateHistogram.min()) / (3.5 * graduationGrateHistogram.std() / n_graduationGrateHistogram**(1/3)))

#Plotting the data

fig, ax = plt.subplots(2, 2, figsize=(15, 12))

sns.histplot(applicationsHistogram, bins=scott_bins_applicationsHistogram, kde=True, color="purple", ax=ax[0,  0])
ax[0, 0].set_title("applications/Frequency")
ax[0, 0].set_xlabel("applications")
ax[0, 0].set_ylabel("Frequency")

sns.histplot(acceptanceHistogram, bins=scott_bins_acceptanceHistogram, kde=True, color="green", ax=ax[0, 1])
ax[0, 1].set_title("Acceptance/Frequency")
ax[0, 1].set_xlabel("Acceptance")
ax[0, 1].set_ylabel("Frequency")   

sns.histplot(expendHistogram, bins=scott_bins_expendHistogram, kde=True, color="blue", ax=ax[1, 0])
ax[1,0].set_title("Expend/Frequency")
ax[1,0].set_xlabel("Expend Rate")
ax[1,0].set_ylabel("Frequency")

sns.histplot(graduationGrateHistogram, bins=scott_bins_graduationGrateHistogram, kde=True, color="red", ax=ax[1, 1])
ax[1,1].set_title("Graduation Rate/Frequency")
ax[1,1].set_xlabel("Graduation Rate")
ax[1,1].set_ylabel("Frequency")   

plt.tight_layout()
plt.show()
```
